```python
from marvin.tools.query import Query
import pandas as pd

from IPython.display import clear_output

from tqdm import tqdm

import sys
```

### Getting Coordinates

```python
df = pd.read_csv('C:/Users/oryan/Documents/PySPAM_Original_Python_MCMC/APySPAM_MCMC/data/all_coords.csv', index_col = 0)
```

```python
df.columns
```

```python
df_prims = (
    df[['Names', 'Prim_RA', 'Prim_DEC']]
    .assign(names_diff = df.Names.apply(lambda x: x + '_1'))
    .rename(columns = {'Prim_RA':'RA', 'Prim_DEC': 'Dec'})
)
df_secs = (
    df[['Names', 'Sec_RA', 'Sec_DEC']]
    .assign(names_diff = df.Names.apply(lambda x: x + '_2'))
    .rename(columns = {'Sec_RA' : 'RA', 'Sec_DEC' : 'Dec'})
)
```

```python
df_comb = pd.concat([df_prims, df_secs]).drop(columns = 'Names')
```

```python
df_comb
```

```python
myfilter = f'radial({df_comb.RA.iloc[0]}, {df_comb.Dec.iloc[0]}, 0.5)'
```

```python
q = Query(search_filter = myfilter)
r = q.run()
```

```python
for i in tqdm(range(len(df_comb))):
    row = df_comb.iloc[i]
    ra = row.RA
    dec = row.Dec
    
    myfilter = f'radial({ra}, {dec}, 0.01)'
    
    q = Query(search_filter = myfilter)
    r = q.run()
    
    if len(r.results) > 0:
        print('Found Data!')
        
        if i == 7 or i == 14 or i == 59 or i == 61 or i == 67: # or i == 68:
            continue
        
        sys.exit()
    else:
        clear_output(wait = True)
```
